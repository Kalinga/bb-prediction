```python
#vid = ["video_0317"]
#bb_cross = []
for v in train_videos:
    anno = imdb._get_annotations(v)
    ped_b = [k for k,v in anno["ped_annotations"].items() if k.endswith('b')]
    for p in ped_b:
        if (1 in anno["ped_annotations"][p]["behavior"]["cross"]):
            #print(v)
            print (p)
            #bb_cross.append((anno["ped_annotations"][p]["bbox"]))
            with open("train/" + p + ".csv", 'w', newline='') as csvfile:
                writer = csv.writer(csvfile,  quoting=csv.QUOTE_MINIMAL)
                for r in (anno["ped_annotations"][p]["bbox"]):
                    print(r)
                    writer.writerow(r)
                csvfile.close()
```
